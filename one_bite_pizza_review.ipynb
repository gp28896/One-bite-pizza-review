```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import urllib.parse
import time

URL = "https://www.onebitepizzarankings.com/"

headers = {
    "User-Agent": "Mozilla/5.0"
}
```

```python
response = requests.get(URL, headers=headers)
```

```python
soup = BeautifulSoup(response.text, "html.parser")
```

```python
# Find table (adjust selector if site changes)
table = soup.find("table")

rows = []
headers = []
```

```python
# Extract headers
for th in table.find_all("th"):
    headers.append(th.text.strip())
```

```python
# Add new column for Google Maps
headers.append("Google Maps Link")

# Extract rows
for tr in table.find_all("tr")[1:]:
    cols = tr.find_all("td")
    if len(cols) == 0:
        continue

    row = [td.text.strip() for td in cols]

    # Build Google Maps link using name + location (if available)
    query = " ".join(row[:2])  # usually name + city
    maps_link = "https://www.google.com/maps/search/?api=1&query=" + urllib.parse.quote(query)

    row.append(maps_link)
    rows.append(row)
```

```python
# Convert to DataFrame
df = pd.DataFrame(rows, columns=headers)
```

```python
print(df)
```

```python
# Save CSV
df.to_csv("one_bite_pizza_rankings.csv", index=False)
```
